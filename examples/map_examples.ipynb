```python
# import sys
# sys.path.append("../")
import pandas as pd
from pyecharts.faker import Faker
from pandasecharts import echart
from pandasecharts.echart import DataFrameEcharts as DFE
```

```python
products = Faker.choose()
dfs = []
for year in (2012, 2013, 2014, 2015):
    dfs.append(
        pd.DataFrame(
            zip(products, Faker.provinces, Faker.values(), Faker.values(), [year]*len(products)),
            columns=["商品", "省份", "商家A", "商家B", "年份"]
        )
    )
df = pd.concat(dfs, axis=0)
```

```python
t_2012: "DFE" = df.query("年份==2012").echart
t_all: "DFE" = df.echart
```

2012年商家A的不同省份的总销量

```python
t_2012.map("省份", "商家A", maptype="china", visualmap=True, agg_func="sum").render_notebook()
```

多个年份商家A的不同省份总销量

```python
t_all.map("省份", "商家A", maptype="china", visualmap=True, agg_func="sum", by="年份").render_notebook()
```

年份作时间轴显示

```python
t_all.map("省份", "商家A", maptype="china", visualmap=True, agg_func="sum", timeline="年份").render_notebook()
```
