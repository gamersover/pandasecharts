```python
# import sys
# sys.path.append("../")
import pandas as pd
from pyecharts.faker import Faker
from pandasecharts import echart
from pandasecharts.echart import DataFrameEcharts as DFE
```

```python
products = Faker.choose()
dfs = []
for year in (2012, 2013, 2014, 2015):
    dfs.append(
        pd.DataFrame(
            zip(products, Faker.values(), Faker.values(), [year]*len(products)),
            columns=["商品", "商家A", "商家B", "年份"]
        )
    )
df = pd.concat(dfs, axis=0)
```

```python
df.head()
```

```python
t_2012: "DFE" = df.query("年份 == 2012").echart
t_all: "DFE" = df.echart
```

2012年商家A的各商品分布

```python
t_2012.boxplot("商家A").render_notebook()
```

2012年所有商家分布

```python
t_2012.boxplot(["商家A", "商家B"]).render_notebook()
```

不同年份商家商品分布

```python
t_2012.boxplot(["商家A", "商家B"], by="年份").render_notebook()
```

将年份作为时间轴显示

```python
t_all.boxplot(["商家A", "商家B"], timeline="年份").render_notebook()
```
