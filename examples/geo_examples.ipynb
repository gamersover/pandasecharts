```python
# import sys
# sys.path.append("../")
import pandas as pd
from pyecharts.faker import Faker
from pandasecharts import echart
from pandasecharts.echart import DataFrameEcharts as DFE
```

```python
products = Faker.choose()
dfs = []
for year in (2012, 2013, 2014, 2015):
    dfs.append(
        pd.DataFrame(
            zip(products, Faker.provinces, Faker.values(), Faker.values(), [year]*len(products)),
            columns=["商品", "省份", "商家A", "商家B", "年份"]
        )
    )
df = pd.concat(dfs, axis=0)
```

```python
t_2012: "DFE" = df.query("年份 == 2012").echart
t_all: "DFE" = df.echart
```

2012年商家A不同省份总销量分布

```python
t_2012.geo("省份", "商家A", agg_func="sum", maptype="china", visualmap=True).render_notebook()
```

显示多个不同商家的省份情况

```python
t_2012.geo("省份", ["商家A", "商家B"], agg_func="sum", maptype="china").render_notebook()
```

按照不同年份显示

```python
t_all.geo("省份", "商家A", agg_func="sum", maptype="china", by="年份", visualmap=True).render_notebook()
```

将年份作为时间轴显示

```python
t_all.geo("省份", "商家A", agg_func="sum", maptype="china", timeline="年份", visualmap=True).render_notebook()
```
